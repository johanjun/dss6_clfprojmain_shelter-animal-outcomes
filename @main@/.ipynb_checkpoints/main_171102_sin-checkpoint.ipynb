```python
def DayuponOutcome(age_str):
    num, unit = age_str.split(' ')
    if unit == 'day' or unit == 'days':
        return(int(num))
    elif age_str == 'week' or age_str == 'weeks':
        return(int(num) * 7 )
    elif age_str == 'month' or age_str == 'months':
        return(int(num) * 30 )
    elif age_str == 'year' or age_str == 'years':
        return(int(num) * 365 )
```

```python
def Nature_Breed(x):
    if '/' in x:
        x = x[0:x.index('/')]
        return x
    elif ' Mix' in x:
        x = x[0:-4]
        return x
    else:
        return x
```
