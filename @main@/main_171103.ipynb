# Color preprocessing

```python
# 'A/B' = 'A' 
def color_del(x):
    if '/' in x:
        x = x[0:x.index('/')]
        return x
    else:
        return x
```

```python
df_pre['Color'] = df_pre['Color'].apply(color_del)
```

```python
df.tail()
```

```python
df_pre.tail()
```

```python
df_pre["Color"].unique()
```

```python
#   
def color_union(x):
    if 'Black' in x:
        x = 'Black'
        return x
    elif 'Blue' in x:
        x = 'Blue'
        return x
    elif 'Red' in x:
        x = 'Red'
        return x
    elif 'Silver' in x:
        x = 'Silver'
        return x
    elif 'Brown' in x:
        x = 'Brown'
        return x
    elif 'Orange' in x:
        x = 'Orange'
        return x
    elif 'Cream' in x:
        x = 'Cream'
        return x
    elif 'Gray' in x:
        x = 'Gray'
        return x
    elif 'Calico' in x:
        x = 'Calico'
        return x
    elif 'Chocolate' in x:
        x = 'Chocolate'
        return x
    elif 'Tortie' in x:
        x = 'Tortie'
        return x
    elif 'Yellow' in x:
        x = 'Yellow'
        return x
    elif 'Liver' in x:
        x = 'Liver'
        return x
    else:
        return x
```

```python
df_pre['Color'] = df_pre['Color'].apply(color_union)
df_pre['Color'].unique()
```

```python
# 40   'etc' 
df_pre['Color'] = df_pre['Color'].replace('Agouti', 'etc')
df_pre['Color'] = df_pre['Color'].replace('Apricot', 'etc')
df_pre['Color'] = df_pre['Color'].replace('Lilac Point', 'etc')
df_pre['Color'] = df_pre['Color'].replace('Pink', 'etc')
df_pre['Color'] = df_pre['Color'].replace('Ruddy', 'etc')
```

```python
df_pre.groupby(['Color']).count()
```
